```python
from faker import Faker
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from num2words import num2words
```

Inserindo a quantidade de estutantes:

```python
Criar = Faker()
Estudante = [[Criar.name(),Criar.random_number(1)] for x in range(10)]
ArquivoF = open('arquivo1.txt','w+')
[ArquivoF.write(f"{x}\n") for x in Estudante]
ArquivoF.close()
print(Estudante)
```

Gráfico:

```python
ArquivoR = open('arquivo1.txt', 'r')
p = ArquivoR.readlines()
db_dados = []
for i in p:
  for j in i:
    if j.isdigit() == True:
      db_dados.append(int(j))

Altura, Comprimento, Colunas = plt.hist(db_dados,facecolor='green',alpha=0.75); 

plt.xlabel('Notas')
plt.ylabel('Probabilidade')
plt.title('Histograma das notas')
plt.grid(True)
plt.show()
```

Wordcloud

```python
N = []

for c in db_dados:
  Numero_W = num2words(c, lang='pt_br')
  N.append(Numero_W)

N = ",".join(N)
Palavras = WordCloud(background_color='yellow',
                           width=900,height=500).generate(N)
plt.imshow(Palavras, interpolation='bilinear')
plt.axis("off")
Palavras.to_file("Nuvem de palavras.png")
```
